Лабораторная работа №1. Расчетная работа по определению уровней акустических сигналов защищаемого помещения
===

```python
import math
from IPython.display import Image
from IPython.core.display import HTML 

def distance(R,D): 
    print("Рассчитываем затухание звука на расстоянии", D, "м, от первоначального значения", R, "дБ ...")
    out=round(R-20*math.log(D,10), 1)
    print("R =", out, "дБ")
    return out

def wall(R,S,K):
    print("Рассчитываем затухание звука за преградой со звукопоглощением", K, "дБ, площадью", S, "м2, от первоначального значения", R,  "дБ ...")
    out=round(R+6-10*math.log(S,10)-K, 1)
    print("R =", out, "дБ")
    return out

def amp(R,K):
    print("Усиление сигнала на", K, "дБ")
    out=R+K
    print("R =", out, "дБ")
    return out

def result(R1,R2):
    if R1>15 and R1-R2<3:
        print("Отсутствует утечка информации")
    else:
        print("Присутствует утечка информации")
```
